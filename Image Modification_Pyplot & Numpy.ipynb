# Flipping the image

```python
#rotate the image by rows/flip the image
plt.imshow(dog[::-1,:,:])
```

# Rotating the image

```python
#rotate/reverse by columns
plt.imshow(dog[:,::-1,:])
```

# Cropping the image 

```python
#crop image by slicing the rows and columns
plt.imshow(dog[0:600,0:800,:])
```

# Blurring the image

```python
#compressing by missing some rows/columns or pixels
plt.imshow(dog[0::20,0::20,:])
```

# Transpose Function

np.transpose()
This function makes rows as columns and columns as rows 
But this works on 2 D arrays/matrices
For 3 D matrix the logic has to be passed to it

(row index, col index, dimension)
(0,1,2)

Now keeping 2 intact interchange the row and col index

(1,0,2) this means row index is 1 and col index is 0

```python
ema_rotate= np.transpose(ema,(1,0,2))
ema_rotate
```

```python
plt.imshow(ema_rotate)
```

```python
ema_crop=ema[200:1000,200:1000,:]
ema_crop
```

```python
plt.imshow(ema_crop)
```

```python
plt.imsave('ema_crop.jpg',ema_crop)
```
